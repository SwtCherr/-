#УЧП гиперболического типа

##Разностные схемы

Одним из примеров УЧП гиперболического типа является линейное однородное уравнение переноса:

$$
\frac{\partial u}{\partial t}+ c\frac{\partial u}{\partial x}=0 \quad t \in[0, \mathrm{~T}], \quad x \in[0, X].
$$

Добавим начальные и граничные условия:
$$
u(0, x) = g(x), \quad u(t, 0)  = f_1(t), \quad u(t, X)  = f_2(t).
$$




Аппроксимировать это уравнение можно используя схемы «явный левый уголок» и «явный правый уголок»:

\begin{equation}
    \dfrac{y_m^{n+1} - y_m^{n}}{\tau} +c \dfrac{y_m^{n} - y_{m-1}^{n}}{h}=0,
\end{equation}


\begin{equation}
    \dfrac{y_m^{n+1} - y_m^{n}}{\tau} +c \dfrac{y_{m+1}^{n} - y_{m}^{n}}{h}=0.
\end{equation}


Обе схемы имеют порядок аппроксимации $O(\tau+h)$.








Запишем разложение в ряд Тейлора:
$$
u(t + \tau, x) \approx u(t, x) + \tau \frac{\partial u}{\partial t}(t, x) + \frac{1}{2} \tau^2 \frac{\partial^2 u}{\partial t^2}(t, x).
$$

Из исходного уравнения следует:
$$
\frac{\partial u}{\partial t} = - c \frac{\partial u}{\partial x}, \quad \frac{\partial^2 u}{\partial t^2} = - c\frac{\partial}{\partial t}  \frac{\partial u}{\partial x} = c^2 \frac{\partial^2 u}{\partial x^2}.
$$

Подставляя это в разложение Тейлора получим:
$$
\mathbf{s}(t+\tau, x) \approx \mathbf{s}(t, x) - \tau \mathbf{A} \frac{\partial \mathbf{s}}{\partial x}(t, x) + \frac{1}{2} \tau^2 \mathbf{A}^2 \frac{\partial^2 \mathbf{s}}{\partial x^2}(t, x).
$$

Для производных по координате используем формулы численного дифференцирования второго порядка:

$$
 \frac{\partial u}{\partial x}(t, x) \approx \frac{u(t, x + h) - u(t, x - h)}{h},
$$

$$
 \frac{\partial ^2u}{\partial x^2}(t, x) \approx \frac{u(t, x - h) - 2u(t, x) + u(t, x + h)}{h^2}.
$$

В результате получаем схему Лакса-Вендроффа:
$$
u_m^{n+1} = u_m^n - \frac{c\tau}{2h}  \left( u_{m+1}^n - u_{m-1}^n \right) + \frac{1}{2} \left( \frac{c\tau}{h} \right)^2  \left( u_{m+1}^n - 2 u_m^n + u_{m-1}^n \right)
$$

По постоению понятно, что схема имеет порядок аппроксимации $O(\tau^2+h^2)$.


##Распространение акустической волны в однородной среде

Часто задачи на распространение волн удобно формулировать в виде системы дифференциальных уравнений первого порядка. Так распространение одномерной акустической волны в однородной среде описывается следующей системой:


$$
\frac{\partial \mathbf{s}}{\partial t} + \mathbf{A} \frac{\partial \mathbf{s}}{\partial x} = \mathbf{0} \quad t \in[0, \mathrm{~T}], \quad x \in[0, X],
$$
$$
\mathbf{s} = \begin{pmatrix}
u \\
p
\end{pmatrix}, \quad \mathbf{A} = \begin{pmatrix}
0 & \dfrac{1}{\rho} \\
\rho c^2 & 0
\end{pmatrix}, \quad \mathbf{0} = \begin{pmatrix}
0 \\
0
\end{pmatrix}.
$$
здесь $u$ -- скорость частиц среды, $p$ -- давление, $\rho$ -- плотность и $c$ -- скорость звука.


Начальные и граничные условия:
$$
u(0, x) = g_1(x), \quad p(0, x) = g_2(x)
$$
$$
\alpha_1 u(t, 0) + \alpha_2 p(t, 0) = f_1(t), \quad \alpha_3 u(t, X) + \alpha_4 p(t, X) = f_2(t).
$$





В таком случае аппроксимация по схеме Лакса-Вендроффа имеет вид:
$$
\mathbf{s}_m^{n+1} = \mathbf{s}_m^n - \frac{\tau}{2h} \mathbf{A} \left( \mathbf{s}_{m+1}^n - \mathbf{s}_{m-1}^n \right) + \frac{1}{2} \left( \frac{\tau}{h} \right)^2 \mathbf{A}^2 \left( \mathbf{s}_{m+1}^n - 2 \mathbf{s}_m^n + \mathbf{s}_{m-1}^n \right)
$$

Покомпонентно с учетом $r = \frac{c\tau}{h}$ получаем:

$$
u_m^{n+1} = u_m^n - \frac{r}{2 c\rho} \left( p_{m+1}^n - p_{m-1}^n \right) + \frac{1}{2} r^2 \left( u_{m+1}^n - 2 u_m^n + u_{m-1}^n \right),
$$
$$
p_m^{n+1} = p_m^n - \frac{\rho c r}{2} \left( u_{m+1}^n - u_{m-1}^n \right) + \frac{1}{2} r^2 \left( p_{m+1}^n - 2 p_m^n + p_{m-1}^n \right).
$$


Решим эту систему при нулевых начальных условиях и граничных условиях следующего вида:

$$
u(0, t) =f(t) = f_0 t (1-t)^n \quad \text{где} \quad f_0 = (1+n)\left(\frac{1+n}{n}\right)^n, \quad n =10,
$$
$$
u(X, t) = g(t) = 0.
$$

Разностная аппроксимаций таких граничных условий следующая:

$$
u_0^{k+1} = f(t_{k+1}), \quad u_N^{k+1} = 0.
$$


Однако для решения задачи нужно получить аппроксимацию второго порядка для граничных условий на давление. Воспользуемся для этого методом фиктивных областей. Вводится фиктивный слой с координатами $(x_{-1}, t_k)$. Тогда справедливо следующее соотношение:

$$
p_0^{k+1} = p_0^k - \frac{r c \rho}{2} \left( u_1^k - u_{-1}^k \right) + \frac{1}{2} r^2 \left( p_1^k - 2 p_0^k + p_{-1}^k \right)
$$

Для исключения фиктивных значений $ u_{-1}^k$ и $p_{-1}^k $ используется приближение:
$$
0.5 \left( u_1^k + u_{-1}^k \right) = u_0^k = f(t_k)
$$

Тогда разностное соотношение принимает вид:
$$
\frac{r}{2 \rho c} \left( p_1^k - p_{-1}^k \right) = u_0^k - u_0^{k-1} = f(t_k) - f(t_{k+1})
$$

В результате получаем формулу для граничного условия:


$$
p_0^{k+1} = p_0^k - \rho c \cdot r \left( u_1^k - f(t_{k+1}) \right) + r^2 \left( p_1^k - p_0^k \right)
$$

Аналогично можно получить выражение для правой границы:

$$
p_N^{k+1} = p_N^k - \rho c \cdot r \left( g(t_{k+1})- u_{N-1}^k \right) + r^2 \left( p_{N-1}^k - p_N^k \right)
$$


В случае, когда на границе задано давление, аналогичная процедура применяется для определения скорости в граничной точке.






```python
import numpy as np
import matplotlib.pyplot as plt

class AcousticWaveSolver:
    def __init__(self, X, T, Nx, Nt, rho, c, u_left_boundary, u_right_boundary):
        self.rho = rho  # Density
        self.c = c  # Speed of sound

        self.X = X  # Total length
        self.T = T  # Total time
        self.Nx = Nx  # Number of spatial intervals
        self.Nt = Nt  # Number of time intervals
        self.dx = X / (Nx - 1)  # Spatial step size
        self.dt = T / (Nt - 1)  # Time step size
        self.r = c * self.dt / self.dx  # Courant number
        self.x = np.linspace(0, X, Nx)  # Spatial grid
        self.t = np.linspace(0, T, Nt)  # Time grid

        # Solution matrices
        self.u = np.zeros((Nt, Nx))
        self.p = np.zeros((Nt, Nx))

        # Set initial conditions
        self.u[0, :] = 0
        self.p[0, :] = 0

        # Set boundary conditions
        self.u_left_boundary = u_left_boundary
        self.u_right_boundary = u_right_boundary

        for k in range(1, self.Nt):
            self.u[k, 0] = self.u_left_boundary(self.t[k])
            self.u[k, -1] = self.u_right_boundary(self.t[k])

    def solve(self):
        for k in range(self.Nt - 1):
            # Set pressure left boundary conditions
            self.p[k + 1, 0] = (self.p[k, 0]
                                - self.r * self.rho * self.c * (self.u[k, 1] - self.u_left_boundary(self.t[k + 1]))
                                + self.r ** 2 * (self.p[k, 1] - self.p[k, 0]))

            # Lax-Wendroff method step
            for i in range(1, self.Nx - 1):
                self.u[k + 1, i] = (self.u[k, i]
                                    - self.r / (2 * self.rho) * (self.p[k, i + 1] - self.p[k, i - 1])
                                    + self.r ** 2 / 2 * (self.u[k, i + 1] - 2 * self.u[k, i] + self.u[k, i - 1]))

                self.p[k + 1, i] = (self.p[k, i]
                                    - self.r * self.rho * self.c / 2 * (self.u[k, i + 1] - self.u[k, i - 1])
                                    + self.r ** 2 / 2 * (self.p[k, i + 1] - 2 * self.p[k, i] + self.p[k, i - 1]))

            # Set pressure right boundary condition
            self.p[k + 1, -1] = (self.p[k, -1]
                                - self.r * self.rho * self.c * (self.u_right_boundary(self.t[k + 1]) - self.u[k, -2])
                                + self.r ** 2 * (self.p[k, -2] - self.p[k, -1]))

    def plot(self, plot_time):
        k = np.abs(self.t - plot_time).argmin()
        plt.figure(figsize=(10, 6))
        plt.plot(self.x, self.u[k, :], label=f'p, t = {self.t[k]:.2f}')
        plt.xlabel('x')
        plt.ylabel('u')
        plt.title(f'Acoustic wave at t = {self.t[k]:.2f}')
        plt.legend()
        plt.grid()
        plt.show()
```

```python
X = 1.0
T = 1.0
Nx = 40
Nt = 60
rho = 1.0
c = 1.0

def u_left_boundary(t):
    n = 10
    f0 = (1 + n) * ((1 + n) / n)**n
    return f0 * t * (1 - t)**n

def u_right_boundary(t):
    return 0

solver = AcousticWaveSolver(X, T, Nx, Nt, rho, c, u_left_boundary, u_right_boundary)
solver.solve()
solver.plot(plot_time=1.0)
```
